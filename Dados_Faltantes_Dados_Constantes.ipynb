Lidar com dados no mundo real, não é uma tarefa muito fácil, precisamos entender e saber lidar com dados ruins. Os dois tipos de dados ruins mais comuns em uma análise de dados são os dados **constantes** e os dados **errôneos**

O **DataSet** Utilizado aqui será referente a casas em São Paulo.

```python
import numpy as np
import pandas as pd
```

```python
casas_sp = pd.read_csv('/houses_sp.csv')
```

```python
casas_sp.head(10)
```

Pelo output podemos ver que as informações de casas disponíveis pertencem somente a cidade de São Paulo, então a coluna 'city' é irrelevante para nós neste contexto, mas vamos avaliar com mais profundidade


```python
casas_sp[('city')].value_counts()
```

```python
casas_sp.info()
```

O **DataSet** possui *7143 entradas* e todas estão no contexto de São Paulo, então podemos remover essa coluna, já que ela não vai interferir nos resultados. Esse é considerado um tipo de dado **"constante"**, e ele é um problema pois pode ocasionar um custo computacional grande nas análises, principalmente se o DataSet contiver alguns milhões de linhas.

```python
casas_sp = casas_sp.drop(columns=['city'])
```

```python
casas_sp.head()
```

Olhando novamente para o **DataSet** podemos perceber também que existe registros denominado "NaN" *(Nulos).* Esses são os dados *faltantes*. E isso é um problema pois se precisarmos calcular uma média, esse dado será desconsiderado e pode influênciar na análise.

```python
casas_sp['rooms']
```

```python
casas_sp.info()
```

Pelo output é possível ver que, na coluna rooms "7092" Registros estão preenchidos.
O restante para chegar na quantidade total estão nulos.

Para tratar esses dados nulos de uma forma onde não perderiamos os outros registros associados a linha dele *(se optassemos por simplesmente excluir)*, seria calcular a mediana da coluna *"rooms"* e preencher  os dados nulos com o resultado dessa mediana

*A média nesses casos não são recomendadas pois elas podem sofrer discrepâncias se houvermos dados muito fora do padrão geral (os outliers)*

```python
casas_sp['rooms'].median()
```

No Geral temos 3 quartos

Agora devemos preencher os dados nulos com uma função do pandas chamada: **fillna()**

A própria função calculou a mediana e preencheu os dados nulos com ela.
Mas precisamos substituir a coluna antiga pela nova tratada, é importante definir a coluna **'rooms'** na variável nova, se não todos os dados serão substituidos.

```python
casas_sp['rooms'] = casas_sp['rooms'].fillna(casas_sp['rooms'].median())
```

```python
casas_sp.info()
```

Podemos ver que agora a **coluna room** possui todas as 7143 linhas preenchidas

Outra coisa que pode vir a acontecer, é o **DataSet** possuir dados que não fazem muito sentido, são os dados **"Errôneos"**

```python
casas_sp.head(10)
```

A coluna floor, tem um registro "-", o que não nos diz nada, mas no contexto dos dados trata-se de uma casa, que não está em nenhum andar, deveria estar 0 no lugar do traço. Mas primeiro vamos olhar a frequência desse dado.

```python
casas_sp['floor'].value_counts()
```

A frequência dele é bem grande e não tem como ser ignorada, precisamos substituir pois se precisarmos fazer calculos *com a média dos andares mais populares por exemplo* isso irá influênciar no resultado.

```python
casas_sp.loc[casas_sp['floor'] == '-', 'floor']
```

```python
casas_sp.loc[casas_sp['floor'] == '-', 'floor'] = '0'
```

```python
casas_sp.loc[casas_sp['floor'] == '0', 'floor']
```

Os **tracinhos** foram substituidos por **zeros**, que representam melhor uma casa que se encontra no térreo.
