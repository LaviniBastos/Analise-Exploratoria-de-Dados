```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

Entendendo um pouco mais a partir da visualização de dados, como as variáveis gerais se relacionam com as variáveis categóricas.
Aqui utiizaremos um dataset da própria biblioteca **seaborn** chamada **íris**. Onde contém informações sobre um conjunto de flores.

```python
iris = sns.load_dataset('iris')
```

```python
iris['species'].value_counts()
```

Temos 3 espécies de flores no **DataSet**

```python
iris.head()
```

A variável categórica desse DataSet é **'species'**. Vamos tentar classificar essa planta baseada nas outrs informações existentes

O gráfico de **histograma** que a a lib '**seaborn**' gera, é um **gráfico de frequência**, que mostra como os valores continuos estão distribuidos.

```python
iris.describe()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x='sepal_length').set_title(f"Distribuição da variável {col}")
```

adicionamos outro parametro na função histplot, que mostra a curva da distribuição

```python
col = 'sepal_length'
sns.histplot(data=iris, x='sepal_length', kde=True).set_title(f"Distribuição da variável {col}")
```

O que nos mostra que existe mais pontos pro lado esquerdo, próximos do 4.5

Podemos adicinar um **for** para visualizarmos mais de um gráfico de uma única vez. Mas para isso vamos usar tbm a **lib matplot** pois ela se comunica com a lib **seaborn**

```python
for col in iris.drop(columns='species'):
  sns.histplot(data=iris, x=col, kde=True).set_title(f"Distribuição da variável {col}")
  plt.show()
```

Agora é preciso fazer uma **interpretação dos gráficos considerando a relação que eles tem com a variável que queremos estudar.** Podemos utilizar o mesmo código anterior, adicionando um outro parâmetro, um parametro de **cor** para visualizar como se distribuem os parâmetros dentro de cada espécie

```python
for col in iris.drop(columns='species'):
  sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f"Distribuição da variável {col}")
  plt.show()
```

Pelos gráficos o **tamanho e a largura da petala** é um critério mais interessante para separar as **espécies**

Outro gráfico que podemos explorar é o barplot..

```python
for col in iris.drop(columns='species'):
  sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y = col)
  plt.show()
```

Esse gráfico basicamente nos mostra uma média, e o tracinho no meio das barras indicam o intervalo de confiança.

Assim como existem diversos outros gráficos bem interessantes a serem explorados nessa biblioteca do Python, que nos ajudam a fazer análises estatisticas mais precisas e ricas com dados contínuos e como podemos associá-los a variáveis categóricas para entermos como essa relação funciona.
