```python
cuisine_rating = pd.read_csv('/Cuisine_rating.csv')
```

```python
cuisine_rating.head()
```

A função head() do pandas nos mostra as 5 primeiras linhas do dataset para que tomemos nota sobre os dados, como eles estão dispostos etc.


```python
cuisine_rating.info()
```

Para identifiicar se o pandas está interpretando os dados corretamente, vendo os tipos dos dados ou se existe valores nulos no dataset, nesse caso não temos valores nulos (o que já nos permite ir para uma próximo passo, que são as perguntas, para extrair respostas do dataset) e temos 200 entradas, para isso usamos a função info(), o que já nos dá uma base para posteriormente limpar ou fazer o tratamento necessário nos dados e ir trabalhando neles.

Uma pergunta mais aberta pra começar a Análise pode ser: **qual tipo de culinária é mais popular na região ou no contexto dos dados fornecidos nesse dataset**

```python
cuisine_rating['Cuisines']
```

Filtramos por "Cuisines", e vamos usar a função values.count(), ela básicamente irá contar a frequência com que aparece cada culinária no dataset

```python
cuisine_rating['Cuisines'].value_counts()
```

Podemos visualizar que a Japonesa é a que mais aparece. Agora podemos ver a quantidade de avaliações somente da culinária Japonesa

```python
cuisine_rating[cuisine_rating['Cuisines'] == 'Japanese']
```

Agora podemos calcular a média das avaliações dessa culinária

```python
cuisine_rating[cuisine_rating['Cuisines'] == 'Japanese']['Overall Rating'].mean()
```

A média das avaliações dessa culinária nessa base é de 3.6.
